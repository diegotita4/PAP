##   <font color = "navy">  Parte 3. Optimización de Sharpe y Evolución del Portafolio. <font color = "black">

+ Con los rendimientos obtenidos en la parte anterior, obtener las ponderaciones eficientes de sharpe, utilizando como ventana de optimización del `01-01-2021` al `12-31-2021`.
    
    
+ Simular el comportamiento de tu portafolio (rendimientos y evolución del dinero) para el período posterior (`01-01-2022` al `12-31-2022`) y presentarlo en un dataframe. Asumir un capital inicial de $1,000,000 de USD.
    

**Productos Esperados**: Ponderaciones óptimas de Sharpe. Dos Data Frames uno con los rendimientos de tu portafolio y otro con la evolución del valor de tu portafolio. Graficar la evolución del portafolio.




```python
# Calcular ponderaciones óptimas de Sharpe
num_assets = len(tickers)
num_portfolios = 100000
risk_free_rate = 0.02  #Cambiar por actual

def portfolio_annualized_performance(weights, mean_returns, cov_matrix):
    returns = np.sum(mean_returns * weights) * 252
    std = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights))) * np.sqrt(252)
    return std, returns

def random_portfolios(num_portfolios, mean_returns, cov_matrix, risk_free_rate):
    results = np.zeros((3,num_portfolios))
    weights_record = []
    for i in range(num_portfolios):
        weights = np.random.random(num_assets)
        weights /= np.sum(weights)
        weights_record.append(weights)
        portfolio_std_dev, portfolio_return = portfolio_annualized_performance(weights, mean_returns, cov_matrix)
        results[0,i] = portfolio_std_dev
        results[1,i] = portfolio_return
        results[2,i] = (portfolio_return - risk_free_rate) / portfolio_std_dev
    return results, weights_record

returns_mean = returns.mean()
returns_cov = returns.cov()
results, weights = random_portfolios(num_portfolios, returns_mean, returns_cov, risk_free_rate)

max_sharpe_idx = np.argmax(results[2])
optimal_weights = weights[max_sharpe_idx]


# Simular comportamiento del portafolio para el período posterior
starting_portfolio_value = 1000000
portfolio_value = [starting_portfolio_value]
portfolio_returns = [0]

for i in range(1, len(df)):
    previous_portfolio_value = portfolio_value[i-1]
    portfolio_return = np.sum(optimal_weights * returns.iloc[i])
    portfolio_value_i = previous_portfolio_value * (1 + portfolio_return)
    portfolio_value.append(portfolio_value_i)
    portfolio_returns.append(portfolio_return)

df_portfolio = pd.DataFrame({'Returns': portfolio_returns, 'Portfolio Value': portfolio_value}, index=df.index)
df_portfolio
```

```python
# Graficar evolución del portafolio
plt.figure(figsize=(8, 6))
plt.plot(df_portfolio['Portfolio Value'])
plt.xlabel('Date')
plt.ylabel('Portfolio Value ($)')
plt.title('Portfolio Evolution')
plt.show()
```

+ Armar un data frame que muestre las siguientes métricas de tu portafolio: Rendimiento Esperado, Volatilidad, Ratio de Sharpe y Alpha de Jensen. Se espera que al menos, el Ratio de Sharpe sea positivo.

**Productos Esperados**: Data Frame con Métricas de Desempeño y comentar los resultados.




* Rendimiento esperado: es la suma de los rendimientos esperados de cada activo ponderados por su peso en el portafolio

* Volatilidad: es la medida de la dispersión de los rendimientos del portafolio. Se puede calcular como la desviación estándar de los rendimientos del portafolio.

* Ratio de Sharpe: es una medida de la rentabilidad ajustada al riesgo del portafolio, que se calcula dividiendo la diferencia entre el rendimiento del portafolio y la tasa libre de riesgo por la volatilidad del portafolio.

$(Rp - Rf) / σp$

* Alpha de Jensen: es una medida de la rentabilidad ajustada al riesgo del portafolio en relación con un benchmark. Se calcula restando la tasa libre de riesgo del rendimiento del portafolio y dividiendo por la sensibilidad del portafolio al benchmark, medida por el coeficiente beta.

$Rp - [Rf + (Rm - Rf) * βp]$

```python
# Calcular métricas del portafolio
portfolio_mean_return = np.sum(optimal_weights * returns_mean) * 252
portfolio_std_dev = np.sqrt(np.dot(optimal_weights.T, np.dot(returns_cov, optimal_weights))) * np.sqrt(252)
portfolio_sharpe_ratio = (portfolio_mean_return - risk_free_rate) / portfolio_std_dev

# Descargar datos del índice SP500
benchmark = yf.download('^GSPC', start="2021-01-01", end="2022-12-31")
benchmark_close = benchmark['Adj Close']
benchmark_returns = benchmark_close.pct_change()

# Calcular alpha de Jensen
beta, alpha = np.polyfit(benchmark_returns[1:], returns.iloc[1:].values, deg=1)

# Crear el DataFrame de métricas del portafolio
metrics = pd.DataFrame({
    'Rendimiento Esperado': portfolio_mean_return,
    'Volatilidad': portfolio_std_dev,
    'Ratio de Sharpe': portfolio_sharpe_ratio,
    'Alpha de Jensen': alpha[1]
}, index=['Portafolio'])

metrics
```
